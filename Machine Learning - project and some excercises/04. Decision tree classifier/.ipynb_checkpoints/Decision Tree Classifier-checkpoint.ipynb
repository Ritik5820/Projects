```python
import numpy as np 
import pandas as pd
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv("salaries.csv")
```

```python
df
```

# Converting text cols to number using dummy function:

```python
salary = df["salary_more_then_100k"]
```

```python
df = df.drop(["salary_more_then_100k"],axis = 1)
```

```python
df
```

```python
dummy = pd.get_dummies(df)
```

```python
dummy
```

# dropping cols due to multicolinearity

```python
dummy = dummy.drop(["company_google","job_sales executive","degree_masters"],axis = 1)
```

```python
dummy
```

# Concatenating dummy and salary together:

```python
dummy
```

```python
salary
```

```python
finaldf = pd.concat((dummy,salary),axis = 1)
```

```python
finaldf
```

# Creating a classifier: 

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
clf = DecisionTreeClassifier()
```

```python
clf.fit(finaldf.drop(["salary_more_then_100k"],axis = 1),finaldf["salary_more_then_100k"])
```

```python
print(clf.score(finaldf.drop(["salary_more_then_100k"],axis = 1),finaldf["salary_more_then_100k"]))
```

# 0,0 for Google | 0,1 for computer programmer | 1 for bachelors

```python
print(clf.predict([[0,0,0,1,1]]))
```
