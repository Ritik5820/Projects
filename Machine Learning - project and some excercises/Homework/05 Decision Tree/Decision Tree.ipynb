# From the titanic.csv file predict who have survived and who didn't on the basis of pclass, sex, age, fare columns

```python
import numpy as np
import pandas as pd 
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv("titanic.csv")
```

```python
df
```

# Removing unneccessary cols: 

```python
df = df.drop(["PassengerId","Name","SibSp","Parch","Ticket","Cabin","Embarked"],axis = 1)
```

```python
df
```

# Checking and filling NA values:

```python
print(df.isnull().sum())
```

```python
print(df["Age"].median())
```

```python
df = df.fillna(28)
```

```python
df
```

```python
print(df.isnull().sum())
```

# Using dummy func to change text values of Sex column:

```python
dummy = pd.get_dummies(df["Sex"])
```

```python
dummy
```

# Removing a col due to multicolinearity: 

```python
dummy = dummy.drop(["male"],axis = 1)
```

```python
dummy
```

# Concatenating df and dummy:

```python
df = df.drop(["Sex"],axis = 1)
```

```python
df
```

```python
finaldf = pd.concat((df,dummy),axis = 1)
```

```python
finaldf
```

# Doing splitting and creating classifier:

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(finaldf.drop(["Survived"],axis = 1),finaldf["Survived"],test_size=0.3)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
clf = DecisionTreeClassifier()
```

```python
clf.fit(x_train,y_train)
```

```python
clf.predict(x_test)
```

```python
print(clf.score(x_test,y_test))
```
