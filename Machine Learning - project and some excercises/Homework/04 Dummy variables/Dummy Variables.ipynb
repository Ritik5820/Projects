# Doing Exploratory Data Analysis:

```python
print(plt.scatter(df["Mileage"],df["Sell Price($)"],marker="+",color = "red"))
print(plt.xlabel("Mileage"))
print(plt.ylabel("Price"))
```

```python
print(plt.scatter(df["Age(yrs)"],df["Sell Price($)"],marker="+",color = "red"))
print(plt.xlabel("Age"))
print(plt.ylabel("Price"))
```

```python
dummy = pd.get_dummies(df["Car Model"])
```

```python
dummy
```

```python
finaldf = pd.concat((df,dummy),axis = 1)
```

```python
finaldf
```

```python
finaldf = finaldf.drop(["Mercedez Benz C class","Car Model"],axis = 1)
```

```python
finaldf
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(finaldf.drop(["Sell Price($)"],axis = 1),finaldf["Sell Price($)"],test_size=0.3)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
model.fit(x_train,y_train)
```

```python
print(model.predict(x_test))
```

```python
print(model.score(x_test,y_test))
```
